```python
movies.columns
#id,title,genre,overview are important columns
```

```python
#creating new dataframes using the above columns
movies=movies[['id','title','overview','genre']]
movies
```

```python
#merge overview and genre in one column
#as for content based recommendation system we need tags
movies['tags']=movies['overview']+movies['genre']
movies
```

```python
#now drop the seperate columns of overview and genre
new_data = movies.drop(columns=['overview','genre'])
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
similarity
```

```python
#we need index of each title based on index we can access movies
new_data[new_data['title']=='The Godfather'].index[0]
```

```python
#list of similarity
distance=sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
  print(new_data.iloc[i[0]].title)
```

```python
def recommend(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance=sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
          print(new_data.iloc[i[0]].title)
```
